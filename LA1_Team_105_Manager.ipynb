```python
from abc import abstractstaticmethod
from contextlib import nullcontext

# creates tabs
tb_all = TabBar(['Staff Login Page', 'Options Page','Manager View', 'Daily Income','Monthly Income', 'Reservation Deep Dive', 'View Booking','Booking Details','View Status','Top Members','Favorite Room Type'])

# a simple function for jumping between tabs
def showPage(tabName):
  with tb_all.output_to(tabName):
    display()

#######################################################
################### Staff Login Page ##################
#######################################################

with tb_all.output_to('Staff Login Page'):
  staff_label = Label ("Staff Login")

  ######################### INPUT BOXES #########################
  # staff username 
  staff_username = widgets.Text(                                                
      value = '',
      placeholder='Insert username',
      description='Username :'
  )
  #staff password
  staff_password = widgets.Password(                                                                            
      value = '',
      placeholder = 'Enter password',
      description = 'Password :'
  )
  #login button
  login_button = widgets.Button(                                                
      description = 'Login',
      button_style = ''
  ) 

  ################# USER VALIDATION: MOVE TO OPTIONS PAGE OR SHOW ERROR ################
  def login_button_clicked(i):
    validation = False
    if len(staff_username.value) == 0 and len(staff_password.value) == 0:
      grid[5,0] = Label("Please fill in the missing information")
    
    elif len(staff_username.value) == 0 and len(staff_password.value) != 0:
      grid[5,0] = Label("Please enter the username")
    
    elif len(staff_username.value) != 0 and len(staff_password.value) == 0:
      grid[5,0] = Label("Please enter a password")
    
    else:
      input = "SELECT STAFF_USERNAME, STAFF_PASSWORD, STAFF_TYPE FROM STAFF WHERE STAFF_USERNAME = '" + staff_username.value + "' AND STAFF_PASSWORD = '" + staff_password.value + "'"
      validation = False
      
      for row in c.execute(input):
        validation = True
        
      if validation:
        c.execute(input)
        checktype = c.fetchall()
        if checktype[0][2] == 'MANAGER':
          grid[5,0] = Label("")
          staff_username.value = ''
          staff_password.value = ''
          showPage('Manager View')
        else:
          grid[5,0] = Label("")
          staff_username.value = ''
          staff_password.value = ''
          showPage('Options Page')  
      else:
        grid[5,0] = Label("Username or Password is wrong, please reenter")

  #run function login_button_clicked
  login_button.on_click(login_button_clicked)

  ############################ GRID ###############################
  grid = GridspecLayout(7, 2, height='300px', width='800px', grid_gap='2px')

  grid[0,0] = widgets.HBox([staff_label])
  grid[1,0] = widgets.HBox([staff_username])
  grid[2,0] = widgets.HBox([staff_password])
  grid[4,0] = widgets.HBox([login_button])
  
  display(grid)


#######################################################
#################### Options Page ####################
#######################################################

with tb_all.output_to('Options Page'):
  options_page = Label ("Select Options Below")

  ############################ INSERT ROOM BUTTON ###############################
  # create insert room button
  insert_room_button = widgets.Button(
      description = "Insert Room",
      width = 'auto',
      height = 'auto',
      button_style= ''
  )

  # create function that moves to page "Insert Room"
  def insert_room_click(i):
    showPage('Insert Room')
  
  # perform "insert_room_click" function when the "insert room" button is clicked
  insert_room_button.on_click(insert_room_click)

  ############################ UPDATE ROOM BUTTON ###############################
  # create update room button
  update_room_button = widgets.Button(
      description = "Update Room",
      button_style = ''
  )

  # create function that moves to page "Update Room Page"
  def update_room_click(i):
    showPage('Update Room Page')

  # perform "update_room_click" function when the "update room" button is clicked
  update_room_button.on_click(update_room_click)

  ############################ UPDATE STATUS BUTTON ###############################
  # create update status button
  update_status_button = widgets.Button(
      description = "Update Status",
      button_style = ''
  )

  # create function that moves to page "Update Status Page"
  def update_status_click(i):
    showPage('Update Status Page')

  # perform "update_status_click" function when the "update status" button is clicked
  update_status_button.on_click(update_status_click)

  ############################ LOG OUT BUTTON ###############################
  # create log out button
  logout_button = widgets.Button(
      description = 'Log Out',
      button_style = ''
  )

  # create function that moves to page "Staff Login Page"
  def logout_click(i):
    showPage('Staff Login Page')

  # perform "logout_click" function when the "Log Out" button is clicked
  logout_button.on_click(logout_click)

  ############################ UPDATE ROOM TYPE BUTTON ###############################
  # create update room button
  update_roomtype_button = widgets.Button(
      description = "Update Room Type",
      button_style = ''
  )

  # create function that moves to page "Update Room Page"
  def update_room_typeclick(i):
    showPage('Update Room Type')

  # perform "update_room_click" function when the "update room" button is clicked
  update_roomtype_button.on_click(update_room_typeclick)

  ############################ GRID ###############################
  options_grid = GridspecLayout(7, 2, height='300px', width='800px', grid_gap='2px')

  options_grid[0,0] = widgets.HBox([options_page])
  options_grid[1,0] = widgets.HBox([insert_room_button])
  options_grid[2,0] = widgets.HBox([update_room_button])
  options_grid[3,0] = widgets.HBox([update_status_button])
  options_grid[4,0] = widgets.HBox([update_roomtype_button])
  options_grid[6,0] = widgets.HBox([logout_button])
  display(options_grid)



#######################################################
##################### Manager View ####################
#######################################################

with tb_all.output_to('Manager View'):
  ########################## LABEL ##########################
  manager_menu_label = Label("Select Dashboard Options Below")

  ########################## BUTTONS ###########################
  # Income button
    ## include the total income per room type
  dailyincome_button = Button(
      description= 'Daily Income')
  monthlyincome_button = Button(
      description= 'Monthly Income')
  totalincome_button = Button(
      description= 'Booking Deep Dive')
  # Room Booking Today
  viewbooking_button = Button(
      description= 'View Booking Details')
  # Top members
  topmembers_button = Button(
      description= 'Top Members')
  # View Booking Status
  viewstatus_button = Button(
      description= 'View Booking Status')
  # Favorite Room Type
  favrtype_button = Button(
      description= 'Favorite Room Type')
  
  # Log out button same as the one on Options Page
  logout_button.on_click(logout_click)

  ########################## FUNCTION ##########################
  # refresh date Datefilter.value = datetime.date.today()
  def showdailyincome(i):
    Datefilter.value = datetime.date.today()
    refreshroomlist_ondaily(i)
    incomedashboard(i)
    showPage('Daily Income')
  
  def showmonthlyincome(i):
    refreshroomlist_onmonthly(i)
    showPage('Monthly Income')
    monthlyincomedashboard(i)
  
  def incomedashboard_page(i):
    allincomedashboard(i)
    showPage('Reservation Deep Dive')

  def viewbookingfortoday(i):
    showPage('View Booking')

  def topmemberpage(i):
  # retrieve data from SQL
    global memberslist, topmemberslist
    c.execute("""
      SELECT *
      FROM
      (SELECT c.CUS_ID, c.CUS_NAME, SUM(RES_NET_BILL)
      FROM RESERVATION r INNER JOIN CUSTOMER c ON r.CUSTOMER_CUS_ID = c.CUS_ID
      WHERE (c.CUS_MEMBERSHIP = 'Yes' or c.CUS_MEMBERSHIP = 'YES') 
      and RES_NET_BILL>0
      GROUP BY c.CUS_ID, c.CUS_NAME
      ORDER BY SUM(RES_NET_BILL) DESC
      ) WHERE ROWNUM <= 5
      """)

    memberslist = c.fetchall()
    topmemberslist(0)
    showPage('Top Members')
  
  def viewstatuspage(i):
    showPage('View Status')

  def favrtype(i):
    showPage('Favorite Room Type')
  
  def goto_home(i):
    showPage('Manager View')

  ###################### BUTTON FUNCTION ######################
  dailyincome_button.on_click(showdailyincome)
  monthlyincome_button.on_click(showmonthlyincome)
  totalincome_button.on_click(incomedashboard_page)
  viewbooking_button.on_click(viewbookingfortoday)
  topmembers_button.on_click(topmemberpage)
  viewstatus_button.on_click(viewstatuspage)
  favrtype_button.on_click(favrtype)

  
  
  ########################## GRIDS ##########################
  managerviewgrid = GridspecLayout(10, 2, height='400px', width='800px', grid_gap='2px')

  managerviewgrid[0,0] = widgets.HBox([manager_menu_label])
  managerviewgrid[1,0] = widgets.HBox([dailyincome_button])
  managerviewgrid[2,0] = widgets.HBox([monthlyincome_button])
  managerviewgrid[3,0] = widgets.HBox([totalincome_button])
  managerviewgrid[4,0] = widgets.HBox([viewbooking_button])
  managerviewgrid[5,0] = widgets.HBox([topmembers_button])
  managerviewgrid[6,0] = widgets.HBox([viewstatus_button])
  managerviewgrid[7,0] = widgets.HBox([favrtype_button])
  managerviewgrid[9,0] = widgets.HBox([logout_button])

  
  display(managerviewgrid)




#######################################################
####################### Daily Income ########################
#######################################################
######## DEFAULT / DAILY INCOME #############################


with tb_all.output_to('Daily Income'):

  ######################### INPUT BOXES #########################
  # By default shows today's income
  Datefilter = DatePicker(value= datetime.date.today())

  # room type list
  def refreshroomlist_ondaily(i):
    global roomtypelist,roomtypelist2,roomtypefilter
    c.execute ("SELECT TYP_NAME, TYP_ID FROM ROOM_TYPE")
    roomtypelist = c.fetchall()
    roomtypelist2 = [("All","All")] + roomtypelist
    roomtypefilter = widgets.Dropdown(options = roomtypelist2,
                                    value = "All")
    incomedashboardgrid[3,1:4] = widgets.HBox([roomtypefilter])
  

  ########################## LABEL ##########################
  # dashboard title
  incomedashboard_label = Label('Income Dashboard',bold=True)
  dailyincome_label = Label('Daily Income')

  # date label 
  date_label = Label("Date :")
  incomedesc_label = Label('Total Income :' + str(Datefilter.value))
  # roomtype label
  roomtypefilter_label = Label("Room Type:")
  

  ########################## BUTTONS ###########################
  incomechart_button = Button(description='Update Chart')
  frdaily_tohome_button = Button(description='Home')

  dailyoption_button = Button(description = 'Daily',disabled = True)
  monthlyoption_button = Button(description = 'Monthly',disabled = False)
  alltime_button = Button(description = 'Deep Dive', disabled = False)
  dailyincomegraph = widgets.Output()
  ########################## FUNCTION ##########################
  dailyincome = []
  datestring = []
  
  # default chart/daily chart
  def incomedashboard(i):
    global dailyincome, datestring, incomedashboardgrid, roomtypefilter, dailyincomegraph
    days = 3
    dailyincome = []
    datestring = []
  
    for day in range(days,0,-1):
      dates = (Datefilter.value-timedelta(day-1)).strftime("%d-%b-%y")
      if roomtypefilter.value == 'All':
        c.execute(
        """ SELECT SUM(RES_NET_BILL/ (RES_CHECK_OUT-RES_CHECK_IN))
        FROM RESERVATION 
        WHERE RES_CHECK_IN <= '{}' AND RES_CHECK_OUT > '{}'
        """
        .format(dates,dates)
        )
      else:
        c.execute(
        """ SELECT SUM(RES_NET_BILL/ (RES_CHECK_OUT-RES_CHECK_IN))
        FROM RESERVATION r
        INNER JOIN CONTAINS ON r.RES_ID = CONTAINS.RESERVATION_RES_ID
        INNER JOIN ROOM ON CONTAINS.ROOM_ROM_ID = ROOM.ROM_ID
        INNER JOIN ROOM_TYPE ON ROOM.ROOM_TYPE_TYP_ID = ROOM_TYPE.TYP_ID
        WHERE RES_CHECK_IN <= '{}' AND RES_CHECK_OUT > '{}' AND ROOM_TYPE.TYP_ID = '{}'
        """
        .format(dates,dates,roomtypefilter.value)
        )
  
      datestring.append(str(dates))
      temp = c.fetchone()[0]
     
      if temp == None :
        dailyincome.append(0)
      else:
        dailyincome.append(round(temp,2))
    
    #visualize the data
    plt.bar(datestring,dailyincome, color= 'turquoise', width = 0.3)  
    for x,y in zip(datestring,dailyincome):
      label = "$ {}".format(y)
      plt.annotate(label, 
                (x,y), 
                textcoords="offset points", 
                xytext=(0,2.5), 
                ha='center')
    plt.ylim(0,)
    plt.xlabel("Date")
    plt.ylabel("Income")
    plt.title("Daily Income")
    
    
    #dailyincomegraph = widgets.Output()
    with dailyincomegraph:
      plt.show()
  
    incomedashboardgrid[5:14,0:] = widgets.HBox([dailyincomegraph])

  ###################### BUTTON FUNCTION ######################
  def updateincomechart(i):
    global dailyincomegraph
    dailyincomegraph.clear_output()
    incomedashboard(i)
  incomechart_button.on_click(updateincomechart)

  def frdaily_tomon(i):
    global dailyincomegraph, today, monthlyincomedashboard
    dailyincomegraph.clear_output()
    refreshroomlist_onmonthly(i)
    #month_dropdown = widgets.Dropdown(options = calendar.month_name[1:13],value = today.strftime('%B'))
    showPage('Monthly Income')
    monthlyincomedashboard(i)
  monthlyoption_button.on_click(frdaily_tomon)

  def frdaily_tohome(i):
    global dailyincomegraph
    dailyincomegraph.clear_output()
    showPage('Manager View')
  frdaily_tohome_button.on_click(frdaily_tohome)

  def frdaily_toall(i):
    global dailyincomegraph
    dailyincomegraph.clear_output()
    allincomedashboard(i)
    showPage('Reservation Deep Dive')

  alltime_button.on_click(frdaily_toall)

  ########################## GRIDS ##########################
  incomedashboardgrid = GridspecLayout(16, 8, height='700px', width='800px', grid_gap='2px')
  #trial = widgets.interact(incomedashboardgrid,incomedashboard)

  incomedashboardgrid[0,0:4] = widgets.HBox([incomedashboard_label])
  incomedashboardgrid[1,0] = widgets.HBox([dailyincome_label])
  
  incomedashboardgrid[2,0] = widgets.HBox([date_label])
  incomedashboardgrid[2,1:4] = widgets.HBox([Datefilter])
  #incomedashboardgrid[2,4:6] = widgets.HBox([incomedesc_label])
  incomedashboardgrid[3,0] = widgets.HBox([roomtypefilter_label])
  incomedashboardgrid[1,4] = widgets.HBox([dailyoption_button])
  incomedashboardgrid[1,5] = widgets.HBox([monthlyoption_button])
  incomedashboardgrid[1,6] = widgets.HBox([alltime_button])
  incomedashboardgrid[15,0] = widgets.HBox([frdaily_tohome_button])
  
  #incomedashboardgrid[4:5,2:7] = widgets.interact(incomedashboard)
  incomedashboardgrid[4,0:3] = widgets.HBox([incomechart_button])
  
  
  display(incomedashboardgrid)





  ###########################################################
  ################ MONTHLY INCOME DASHBOARD #################
  ###########################################################


with tb_all.output_to('Monthly Income'):
  Mon_incomedashboardgrid = GridspecLayout(18, 8, height='800px', width='800px', grid_gap='2px')
  ########################## LABEL AND ALL INPUT ##########################
  frmonth_tohome_button = Button(description='Home')
  monthlyincomedashboard_label = Label('2022 Monthly Income')
  wait_label = Label('Calculating... please wait')
  frmon_tohome_button = Button(description='Home')

  today = datetime.date.today()
  
  frmon_dailyoption_button = Button(description = 'Daily',disabled = False)
  frmon_monthlyoption_button = Button(description = 'Monthly',disabled = True)
  frmon_alltime_button = Button(description = 'Deep Dive', disabled = False)
  monthlyincomechart_button = Button(description='Update Chart')
  monthincome = widgets.Output()

  month_label = Label('Month :')   
  month_dropdown = widgets.Dropdown(options = calendar.month_name[1:13],value = today.strftime('%B'))
  year = today.year
  monthly_income =[]
  monthlists = []
  
  # refresh room list
  def refreshroomlist_onmonthly(i):
    global mon_roomtypelist,mon_roomtypelist2,mon_roomtypefilter
    c.execute ("SELECT TYP_NAME, TYP_ID FROM ROOM_TYPE")
    mon_roomtypelist = c.fetchall()
    mon_roomtypelist2 = [("All","All")] + mon_roomtypelist
    mon_roomtypefilter = widgets.Dropdown(options = mon_roomtypelist2,
                                    value = "All")
    Mon_incomedashboardgrid[3,1:4] = widgets.HBox([mon_roomtypefilter])


  # monthly income dashboard function
  def monthlyincomedashboard(i):
    global income, datestring, Mon_incomedashboardgrid, mon_roomtypefilter,month_dropdown, today, wait_label, monthincome, monthlists, monthly_income,year
    # show wait label to show function is calculating
    Mon_incomedashboardgrid[4,3:5] = widgets.HBox([wait_label])
    
    # determine the month and year of today or selected month value
    month = datetime.datetime.strptime(month_dropdown.value, "%B").month
    month_startdate = month
    year = today.year
    year_startdate = year

    # limit the amount of monthly income displayed
    totalmonths=3

    # declare some variable to zero or empty
    monthly_income =[]
    monthlists = []
    
    # month name on a list with the most recent at the left
    for i in range(totalmonths):
      monthlists.append(calendar.month_name[month] + " " + str(year))
      month-=1
      if month == 0:
        year-=1
        month = 12
    
    monthlists.reverse()
    # declare start date
    start_date = datetime.datetime(year_startdate,month_startdate,1)
    # calculate days in a month
    daysInMonth = calendar.monthrange(year_startdate,month_startdate)[1]

    # sum loop
    for month in range(totalmonths,0,-1):
      income = 0
      # for each day in a particular month, sum up the daily income and convert it to monthly
      for single_date in (start_date + timedelta(n) for n in range(daysInMonth)):
    
        everydate = single_date.strftime("%d-%b-%y")
        if mon_roomtypefilter.value == 'All':
          c.execute("""SELECT SUM(RES_NET_BILL/ (RES_CHECK_OUT-RES_CHECK_IN)) FROM RESERVATION 
          WHERE RES_CHECK_IN <= '{}' AND RES_CHECK_OUT > '{}'"""
          .format(everydate,everydate))
        else:
          c.execute(
          """ SELECT SUM(RES_NET_BILL/ (RES_CHECK_OUT-RES_CHECK_IN))
          FROM RESERVATION r
          INNER JOIN CONTAINS ON r.RES_ID = CONTAINS.RESERVATION_RES_ID
          INNER JOIN ROOM ON CONTAINS.ROOM_ROM_ID = ROOM.ROM_ID
          INNER JOIN ROOM_TYPE ON ROOM.ROOM_TYPE_TYP_ID = ROOM_TYPE.TYP_ID
          WHERE RES_CHECK_IN <= '{}' AND RES_CHECK_OUT > '{}' AND ROOM_TYPE.TYP_ID = '{}'
          """
        .format(everydate,everydate,mon_roomtypefilter.value)
        )
        temp = c.fetchone()[0]
        if temp != None :
          income += temp
        
        # insert to a list of monthly income
      monthly_income.append(round(income,2))
      

      # change the current month to previous month
      month_startdate -=1
      # if current month is january, change the year to previous year and month to december
      if month_startdate ==0:
        year_startdate-=1
        month_startdate=12
      daysInMonth = calendar.monthrange(year_startdate,month_startdate)[1]
      start_date = datetime.datetime(year_startdate,month_startdate,1)
      
    monthly_income.reverse()
    #visualize the data
    plt.bar(monthlists,monthly_income, color= 'turquoise', width = 0.4) 

    for x,y in zip(monthlists,monthly_income):
      label = "$ {}".format(y)
      plt.annotate(label, 
                    (x,y), 
                    textcoords="offset points", 
                    xytext=(0,2.5), 
                    ha='center')
    
    plt.ylim(0)
    plt.xlabel("Month")
    plt.ylabel("Income")
    plt.title("Monthly Income")

    with monthincome:
      plt.show()
      Mon_incomedashboardgrid[4,3:5] = widgets.HBox([Label('')])
    Mon_incomedashboardgrid[5:16,0:] = widgets.HBox([monthincome])
    
  
  
  ###################### BUTTON FUNCTION ######################
  def updatemonthlyincomechart(i):
    global monthincome
    monthincome.clear_output()
    monthlyincomedashboard(i)
  monthlyincomechart_button.on_click(updatemonthlyincomechart)

  def frmon_todaily(i):
    global monthly_income, Datefilter, monthincome
    Datefilter.value = datetime.date.today()
    refreshroomlist_ondaily(i)
    incomedashboard(i)
    monthincome.clear_output()
    showPage('Daily Income')
  
  def frmon_tohome(i):
    global monthincome
    monthincome.clear_output()
    showPage('Manager View')

  def frmon_toall(i):
    global monthincome
    monthincome.clear_output()
    allincomedashboard(i)
    showPage('Reservation Deep Dive')
    
  frmon_dailyoption_button.on_click(frmon_todaily)
  frmon_alltime_button.on_click(incomedashboard_page)
  frmon_tohome_button.on_click(frmon_tohome)
  frmon_alltime_button.on_click(frmon_toall)

  ########################## GRIDS ##########################
  
  #trial = widgets.interact(incomedashboardgrid,incomedashboard)

  Mon_incomedashboardgrid[0,0:4] = widgets.HBox([incomedashboard_label])
  Mon_incomedashboardgrid[1,0:3] = widgets.HBox([monthlyincomedashboard_label])
  
  Mon_incomedashboardgrid[2,0] = widgets.HBox([month_label])
  Mon_incomedashboardgrid[2,1:4] = widgets.HBox([month_dropdown])
  #incomedashboardgrid[2,4:6] = widgets.HBox([incomedesc_label])
  Mon_incomedashboardgrid[3,0] = widgets.HBox([roomtypefilter_label])
  Mon_incomedashboardgrid[1,4] = widgets.HBox([frmon_dailyoption_button])
  Mon_incomedashboardgrid[1,5] = widgets.HBox([frmon_monthlyoption_button])
  Mon_incomedashboardgrid[1,6] = widgets.HBox([frmon_alltime_button])
  # Mon_incomedashboardgrid[3,1:4] = widgets.HBox([mon_roomtypefilter])
  #incomedashboardgrid[4:5,2:7] = widgets.interact(incomedashboard)
  Mon_incomedashboardgrid[4,0:2] = widgets.HBox([monthlyincomechart_button])
  Mon_incomedashboardgrid[17,0] = widgets.HBox([frmon_tohome_button])
 
  
  display(Mon_incomedashboardgrid)




#######################################################
################### RESERVATION DETAILS ###################
#######################################################
# Drill down income per room type
with tb_all.output_to('Reservation Deep Dive'):
  ########################## LABEL AND INPUT ##########################
  totalincome_label = Label('Deep Dive on Reservation')
  options_label = Label('Options :')
  detailslist = ['Reward', 'Customer','Pet Species', 'Discount']
  dashboarddetailsoptions = widgets.Dropdown(options = detailslist)
  totalincome_details = Label('')
  alltime_incomedashboardgrid = GridspecLayout(14, 8, height='700px', width='800px', grid_gap='2px')
  deepdive = widgets.Output()

  ######################### BUTTONS ###########################
  frall_dailyoption_button = Button(description = 'Daily',disabled = False)
  frall_monthlyoption_button = Button(description = 'Monthly',disabled = False)
  frall_alltime_button = Button(description = 'Deep Dive', disabled = True)
  frall_tohome_button = Button(description='Home')
  
  x=[]
  y=[]
  
  
  ###################### BUTTON FUNCTION ######################
  def onoptionpick(i):
    allincomedashboard(i['new'])
  def allincomedashboard(i):
    global dashboarddetailsoptions,deepdive_details, x, y,alltime_incomedashboardgrid
    
    deepdive.clear_output()

    x = []
    y = []
    if dashboarddetailsoptions.value == "Reward":
      deepdive_details = Label('Total Claim per ' + dashboarddetailsoptions.value)
      alltime_incomedashboardgrid[2,0:4] = widgets.HBox([deepdive_details])
      c.execute(
        """SELECT r.REWARD_ID, r.REWARD_NAME, COUNT(REWARDS_CLAIMED.CLA_QUANTITY)
        FROM REWARD r
        INNER JOIN REWARDS_CLAIMED ON r.REWARD_ID = REWARDS_CLAIMED.REWARD_REWARD_ID
        GROUP BY r.REWARD_ID, r.REWARD_NAME
        ORDER BY COUNT(REWARDS_CLAIMED.CLA_QUANTITY) DESC
        """
        )
    elif dashboarddetailsoptions.value == "Customer":
      deepdive_details = Label('Total Income per ' + dashboarddetailsoptions.value)
      alltime_incomedashboardgrid[2,0:4] = widgets.HBox([deepdive_details])
      c.execute("""
      SELECT c.CUS_ID, c.CUS_NAME, SUM(RES_NET_BILL)
      FROM RESERVATION r INNER JOIN CUSTOMER c ON r.CUSTOMER_CUS_ID = c.CUS_ID
      WHERE RES_NET_BILL>0
      GROUP BY c.CUS_ID, c.CUS_NAME
      ORDER BY SUM(RES_NET_BILL) DESC
      """)
    elif dashboarddetailsoptions.value == "Pet Species":
      deepdive_details = Label('Total Pet per ' + dashboarddetailsoptions.value)
      alltime_incomedashboardgrid[2,0:4] = widgets.HBox([deepdive_details])
      c.execute(
        """SELECT p.PET_SPECIES, p.PET_SIZE, COUNT(i.PET_PET_ID)
        FROM PET p
        INNER JOIN IS_INCLUDED i ON p.PET_ID = i.PET_PET_ID
        INNER JOIN RESERVATION r ON i.RESERVATION_RES_ID = r.RES_ID
        GROUP BY PET_SPECIES, p.PET_SIZE 
        ORDER BY COUNT(i.PET_PET_ID) DESC
        """
        )
    elif dashboarddetailsoptions.value == "Discount":
      deepdive_details = Label('Popular Promotion')
      alltime_incomedashboardgrid[2,0:4] = widgets.HBox([deepdive_details])
      c.execute(
        """SELECT d.DIS_ID, d.DIS_CODE, COUNT(DISCOUNT_CODE_DIS_ID)
        FROM DISCOUNT_CODE d
        INNER JOIN RESERVATION r ON d.DIS_ID = r.DISCOUNT_CODE_DIS_ID
        GROUP BY d.DIS_ID, d.DIS_CODE
        ORDER BY COUNT(DISCOUNT_CODE_DIS_ID) DESC
        """
        )

    result = c.fetchall()

    for i in result:
      if dashboarddetailsoptions.value == "Pet Species":
        x.append(i[1]+ ' '+i[0])
        y.append(i[2])
      else:
        x.append(i[1])
        y.append(i[2])
      
    
    #visualize data
    plt.bar(x,y, color= 'turquoise', width = 0.4) 
    plt.bar(x,y, color= 'turquoise', width = 0.4) 

    for x,y in zip(x,y):
        label = "{}".format(y)
        plt.annotate(label, 
                    (x,y), 
                    textcoords="offset points", 
                    xytext=(0,2.5), 
                    ha='center')

    plt.ylim(0)

    if dashboarddetailsoptions.value == "Reward":
      plt.xlabel("Reward Name")
      plt.ylabel("Total Claimed")
      plt.title("Total Rewards Claimed")
    elif dashboarddetailsoptions.value == "Customer":
      plt.xlabel("Customer")
      plt.ylabel("Total Income")
      plt.title("Total Income per Customer ($)")
    elif dashboarddetailsoptions.value == "Pet Species":
      plt.xlabel("Pet Species")
      plt.ylabel("Total Booking")
      plt.title("Total Booking per Pet Species")
    else:
      plt.xlabel("Discount Name")
      plt.ylabel("Total Claimed")
      plt.title("Popular Discount Code")

    #visualize the data
    with deepdive:
      plt.show()
    
    alltime_incomedashboardgrid[3:12,0:5] = widgets.HBox([deepdive])
    

  def frall_tomon(i):
    global dailyincomegraph, today, monthlyincomedashboard
    deepdive.clear_output()
    dailyincomegraph.clear_output()
    refreshroomlist_onmonthly(i)
    showPage('Monthly Income')
    monthlyincomedashboard(i)
 
  def frall_todaily(i):
    global monthly_income, Datefilter, monthincome
    Datefilter.value = datetime.date.today()
    refreshroomlist_ondaily(i)
    incomedashboard(i)
    deepdive.clear_output()
    showPage('Daily Income')
  
  def frall_tohome(i):
    deepdive.clear_output()
    showPage('Manager View')
  
  

  dashboarddetailsoptions.observe(onoptionpick,names='value')
  frall_dailyoption_button.on_click(frall_todaily)
  frall_monthlyoption_button.on_click(frall_tomon)
  frall_tohome_button.on_click(frall_tohome)


  ########################## GRIDS ##########################
  alltime_incomedashboardgrid[0,0:4] = widgets.HBox([totalincome_label])

  alltime_incomedashboardgrid[1,0] = widgets.HBox([options_label])
  alltime_incomedashboardgrid[1,1:4] = widgets.HBox([dashboarddetailsoptions])
  alltime_incomedashboardgrid[1,4] = widgets.HBox([frall_dailyoption_button])
  alltime_incomedashboardgrid[1,5] = widgets.HBox([frall_monthlyoption_button])
  alltime_incomedashboardgrid[1,6] = widgets.HBox([frall_alltime_button])
  alltime_incomedashboardgrid[13,0] = widgets.HBox([frall_tohome_button])

  display(alltime_incomedashboardgrid)






###############################################################
######################## View Booking #########################
###############################################################

with tb_all.output_to('View Booking'):
  viewbookingdetailsgrid = GridspecLayout(14, 5, height='600px', width='800px', grid_gap='2px')
  
  #declare date and variable
  currentdate = datetime.date.today()
  Datefilter_viewbooking = DatePicker(value= currentdate)

  #today = currentdate.strftime("%d-%b-%y")
  
  
  ########################## BUTTONS ###########################
  

  nextbooking_button = widgets.Button(description='Next')
  previousbooking_button = widgets.Button(description='Previous')
  viewbooking_button = widgets.Button(description='View Booking')
  managerhome_button = widgets.Button(description='Home')

  ########################## FUNCTION ##########################
  bookingid = []
  currentbookingid = []
  count = 0
  currentpage = 0
  limitperpage = 5

  booking_button = widgets.RadioButtons(
      options = currentbookingid
     )


  def retrieve_reservation_data(today):
    global count, bookingid, currentbookingid, viewbookingdetailsgrid, totalbooking_label, booking_button, currentpage, limitperpage
    # retrieve reservation data

    c.execute(
    """ SELECT RES_ID
    FROM RESERVATION 
    WHERE RES_CHECK_IN = '{}'
    """
    .format(today.strftime("%d-%b-%y"))
    )
    
    allbooking = c.fetchall()

    bookingid= []
    currentbookingid = []
    currentpage = 0
    count=0
    for x in allbooking:
      bookingid.append(x[0])
      count +=1
    
    if count == 0:
      viewbookingdetailsgrid[4,0:3] = widgets.HBox([Label('')])
      viewbookingdetailsgrid[5:8,0:2] = widgets.HBox([Label('No Booking Recorded')])
      #viewbookingdetailsgrid[5:7,0].layout = widgets.Layout(display = 'none')
    else: 
      selectbookingbelow_label = Label('Please select from the booking below:')
      viewbookingdetailsgrid[4,0:3] = widgets.HBox([selectbookingbelow_label])
      currentbookingid = bookingid[0:limitperpage]
      booking_button = widgets.RadioButtons(
      options = currentbookingid
      )
      viewbookingdetailsgrid[5:8,0:2] = widgets.HBox([booking_button])
      viewbookingdetailsgrid[10,1:3] = widgets.HBox([Label('')])

    if count > limitperpage:
      viewbookingdetailsgrid[9,3] = widgets.HBox([nextbooking_button])
    

      # put data on radio button
   

    ########################## LABEL ##########################
    bookingfortoday_label = Label('Booking for ' + today.strftime("%d %B %Y"))
    totalbooking_label = Label('Total booking : ' + str(count))
    viewbookingdetailsgrid[3,0] = widgets.HBox([totalbooking_label])
    viewbookingdetailsgrid[2,0:3] = widgets.HBox([bookingfortoday_label])
    
    
  def ondatepick(i):
    retrieve_reservation_data(i['new'])


  retrieve_reservation_data(currentdate)
  
  Datefilter_viewbooking.observe(ondatepick, names="value")

  viewbooking_label = Label('View Bookings')

  ########################## BUTTONS ###########################
  

  nextbooking_button = widgets.Button(description='Next')
  previousbooking_button = widgets.Button(description='Previous')
  viewbooking_button = widgets.Button(description='View Booking')
  managerhome_button = widgets.Button(description='Home')



########################### BUTTON FUNCTION ######################
  def nextbutton(i):
    global currentpage, limitperpage, currentbookingid, bookingid, booking_button

    currentpage +=1
    currentbookingid = bookingid[currentpage*limitperpage:(currentpage+1)*limitperpage]
    booking_button = widgets.RadioButtons(
      options = currentbookingid
    )
    viewbookingdetailsgrid[5:8,0:2] = widgets.HBox([booking_button])

    viewbookingdetailsgrid[9,0:2] = widgets.HBox([previousbooking_button])

    if count < ((currentpage+2)*limitperpage):
      viewbookingdetailsgrid[9,3].layout = widgets.Layout(visibility = 'hidden')


  def previousbutton(i):
    global currentpage, limitperpage, currentbookingid, bookingid, booking_button

    currentpage -=1
    currentbookingid = bookingid[currentpage*limitperpage:(currentpage+1)*limitperpage]
    booking_button = widgets.RadioButtons(
      options = currentbookingid
    )
    viewbookingdetailsgrid[5:8,0:2] = widgets.HBox([booking_button])

    if currentpage >0:
      viewbookingdetailsgrid[9,0:2] = widgets.HBox([previousbooking_button])
    else:
      viewbookingdetailsgrid[9,0:2].layout = widgets.Layout(visibility = 'hidden')
      viewbookingdetailsgrid[9,3] = widgets.HBox([nextbooking_button])  
      

  def viewbookingdetails(i):
    global currentbookingid, bookingid, booking_button, count
    
    if count == 0:
      warningmessage_label = Label('There is no booking to retrieve')
      viewbookingdetailsgrid[10,1:3] = widgets.HBox([warningmessage_label])
    else:
      viewbookingdetailsgrid[10,1:3] = widgets.HBox([Label('')])
      viewdetails(0)
      showPage('Booking Details')
      
  
  def fromviewdetails_tohome(i):
    viewbookingdetailsgrid[9,1:3] = widgets.HBox([Label('')])
    showPage('Manager View')


############################## GRIDS ##########################

    #trial = widgets.interact(incomedashboardgrid,incomedashboard)


  viewbookingdetailsgrid[0,0] = widgets.HBox([viewbooking_label])
  viewbookingdetailsgrid[1,0] = widgets.HBox([Label('Date :')])
  viewbookingdetailsgrid[1,1:3] = widgets.HBox([Datefilter_viewbooking])
  
  #viewbookingdetailsgrid[3,0] = widgets.HBox([totalbooking_label])

  #viewbookingdetailsgrid[2,0:3] = widgets.HBox([bookingfortoday_label])
  viewbookingdetailsgrid[5:8,0:2] = widgets.HBox([booking_button])
  if count == 0:
    viewbookingdetailsgrid[5:8,0:2].layout = widgets.Layout(display = 'none')
    viewbookingdetailsgrid[5:8,0:2] = widgets.HBox([Label('No Booking For Today')])

  if count > limitperpage:
    viewbookingdetailsgrid[9,3] = widgets.HBox([nextbooking_button])

  viewbookingdetailsgrid[10,0] = widgets.HBox([viewbooking_button])
  viewbookingdetailsgrid[11,0] = widgets.HBox([managerhome_button])

  display(viewbookingdetailsgrid)

  previousbooking_button.on_click(previousbutton)
  nextbooking_button.on_click(nextbutton)
  viewbooking_button.on_click(viewbookingdetails)
  managerhome_button.on_click(fromviewdetails_tohome)



###############################################################
###################### Booking Details ########################
###############################################################

with tb_all.output_to('Booking Details'):
  currentdate = datetime.date.today()
  #todaystring = currentdate.strftime("%d-%b-%y")
  bookingforres_label = Label('Booking details of ' + str(booking_button.value))

  ########################## FUNCTION ##########################
  
  def viewdetails(i):
    global booking_button, bookingdetailsgrid,bookingforres_label
    
    bookingforres_label = Label('Booking details of ' + str(booking_button.value))
    # retrieve reservation data
    c.execute("""
      SELECT r.RES_CHECK_IN, r.RES_CHECK_OUT, r.RES_NET_BILL, 
      r.RES_NIGHTS, c.CUS_NAME, c.CUS_PHONE, c.CUS_MEMBERSHIP
      FROM RESERVATION r
      LEFT JOIN CUSTOMER c ON r.CUSTOMER_CUS_ID = c.CUS_ID
      WHERE RES_ID = '{}'
      """
      .format(booking_button.value)
      )

    reservationdata = c.fetchone()


    ########################## LABEL ##########################
    #result label
    r_checkin = Label(reservationdata[0].strftime("%d %B %Y"))
    r_checkout = Label(reservationdata[1].strftime("%d %B %Y"))
    r_netbill = Label(str(reservationdata[2]))
    r_resnights = Label(str(reservationdata[3]))
    
    
    r_cusname = Label(reservationdata[4])
    r_cusphone = Label(reservationdata[5])
    if reservationdata[6] == 'No':
      r_member = Label('Non-Member')
    else:  
      r_member = Label('Member')

    c.execute("""
         SELECT rtype.TYP_NAME 
         FROM RESERVATION r,CONTAINS c, ROOM r, ROOM_TYPE rtype 
         WHERE r.RES_ID = '{}' and r.RES_ID = c.RESERVATION_RES_ID and
         c.ROOM_ROM_ID = r.ROM_ID and
         r.ROOM_TYPE_TYP_ID = rtype.TYP_ID
         """
         .format(booking_button.value)
     )
    result = c.fetchone()
    r_rtype = Label(result[0])
  
    
    ########################## GRIDS ##########################

    bookingdetailsgrid[2,1:3] = widgets.HBox([r_checkin])
    bookingdetailsgrid[3,1:3] = widgets.HBox([r_checkout])

    bookingdetailsgrid[4,1:3] = widgets.HBox([r_resnights])
    bookingdetailsgrid[5,1:3] = widgets.HBox([r_rtype])
    #bookingdetailsgrid[6,1] = widgets.HBox([bd_petname])
    bookingdetailsgrid[6,1:3] = widgets.HBox([r_netbill])
    bookingdetailsgrid[9,1:3] = widgets.HBox([r_cusname])
    bookingdetailsgrid[10,1:3] = widgets.HBox([r_cusphone])
    bookingdetailsgrid[11,1:3] = widgets.HBox([r_member])
    bookingdetailsgrid[0,0:2] = widgets.HBox([bookingforres_label])
    bookingdetailsgrid[14,0] = widgets.HBox([bd_viewbooking_button])
    bookingdetailsgrid[15,0] = widgets.HBox([detailstohome_button])
    
  ################## TEMPLATE FORMAT ###########################
  bd_checkindate = Label('Check In Date :')
  bd_checkoutdate = Label('Check Out Date :')
  bd_resnights = Label('Total nights of stay :')
  bd_roomtype = Label('Room Type :')
  #bd_petname = Label('Total Pet(s) :')
  bd_netbill = Label('Total Bill (after discount) :')
  

  bd_customer=Label('Customer Details')
  bd_cusname = Label('Customer Name :')
  bd_cusphone = Label('Phone No. :')
  bd_cusmember = Label('Customer Status :')
  
  ########################## GRIDS ##########################
  bookingdetailsgrid = GridspecLayout(16, 5, height='600px', width='800px', grid_gap='2px')
  bookingdetailsgrid[0,0:2] = widgets.HBox([bookingforres_label])

  bookingdetailsgrid[2,0] = widgets.HBox([bd_checkindate])
  bookingdetailsgrid[3,0] = widgets.HBox([bd_checkoutdate])

  bookingdetailsgrid[4,0] = widgets.HBox([bd_resnights])
  bookingdetailsgrid[5,0] = widgets.HBox([bd_roomtype])
  #bookingdetailsgrid[6,0] = widgets.HBox([bd_petname])
  bookingdetailsgrid[6,0] = widgets.HBox([bd_netbill])
  bookingdetailsgrid[8,0] = widgets.HBox([bd_customer])
  bookingdetailsgrid[9,0] = widgets.HBox([bd_cusname])
  bookingdetailsgrid[10,0] = widgets.HBox([bd_cusphone])
  bookingdetailsgrid[11,0] = widgets.HBox([bd_cusmember])
  
  display(bookingdetailsgrid)

  def backtoview(i):
    bookingdetailsgrid[2,1:3] = widgets.HBox([Label('')])
    bookingdetailsgrid[3,1:3] = widgets.HBox([Label('')])
    
    bookingdetailsgrid[4,1:3] = widgets.HBox([Label('')])
    bookingdetailsgrid[5,1:3] = widgets.HBox([Label('')])
    #bookingdetailsgrid[6,1] = widgets.HBox([bd_petname])
    bookingdetailsgrid[6,1:3] = widgets.HBox([Label('')])
    bookingdetailsgrid[9,1:3] = widgets.HBox([Label('')])
    bookingdetailsgrid[10,1:3] = widgets.HBox([Label('')])
    bookingdetailsgrid[11,1:3] = widgets.HBox([Label('')])
    showPage('View Booking')
  
  def frdetails_tohome(i):
    backtoview(i)
    showPage('Manager View')
  ######################### INPUT BOXES #########################
  ########################## BUTTONS ###########################

  bd_viewbooking_button = widgets.Button(description='Back to View Booking')
  detailstohome_button = widgets.Button(description='Home')

  bd_viewbooking_button.on_click(backtoview)

  detailstohome_button.on_click(frdetails_tohome)



#######################################################
################### TOP 5 MEMBERS #####################
#######################################################

with tb_all.output_to('Top Members'):
  ########################## LABEL ##########################
  topmembersgrid = GridspecLayout(17, 5, height='600px', width='800px', grid_gap='2px')
  topmembers_label=Label("Top 5 Members based on Total Spent")
  
  
  topmemberdashboard = widgets.Output()
  frmembers_tohome_button = widgets.Button(description='Home')

  membername = []
  memberincome = []

  def topmemberslist(i):
    global memberslist, topmembersgrid
    top_members = len(memberslist)
    for x in range(top_members):
      topmembersgrid[x+1,0] = Label(str(x+1)+". " + memberslist[x][1])
      membername.append(memberslist[x][1])
      memberincome.append(memberslist[x][2])

    plt.bar(membername,memberincome, color= 'turquoise', width = 0.4) 

    for x,y in zip(membername,memberincome):
      label = "$ {}".format(y)
      plt.annotate(label, 
                  (x,y), 
                  textcoords="offset points", 
                  xytext=(0,2.5), 
                  ha='center')

      plt.ylim(0)
      plt.xlabel("Member")
      plt.ylabel("Total Spend")
      plt.title("Top Members based on Total Spent")

    # Home button
    

    #visualize the data
    with topmemberdashboard:
      plt.show()


    
    topmembersgrid[7:15,0:5] = widgets.HBox([topmemberdashboard])
  frmembers_tohome_button.on_click(goto_home)
  topmembersgrid[0,0:3] = widgets.HBox([topmembers_label])
  topmembersgrid[16,0] = widgets.HBox([frmembers_tohome_button])
  
  display(topmembersgrid)
  
  

#######################################################
################### Favorite Room Type ##################
#######################################################
with tb_all.output_to('Favorite Room Type'):
  rtypegrid = GridspecLayout(17, 5, height='600px', width='800px', grid_gap='2px')
  frfav_tohome_button = widgets.Button(description='Home')
  favroomtypedashboard = widgets.Output()
  frfav_tohome_button.on_click(goto_home)
  
  ########################## LABEL ##########################
  favrtype_label = Label('Favorite Room Type Based on Number of Booking')


  def showfavrtype(i):
    c.execute("""
      SELECT *
      FROM
      (
      SELECT rtype.TYP_ID, rtype.TYP_NAME, COUNT(r.RES_ID), SUM(RES_NET_BILL)
      FROM RESERVATION r 
      INNER JOIN CONTAINS c ON r.RES_ID = c.RESERVATION_RES_ID
      INNER JOIN ROOM rm ON c.ROOM_ROM_ID = rm.ROM_ID
      INNER JOIN ROOM_TYPE rtype ON rm.ROOM_TYPE_TYP_ID = rtype.TYP_ID 
      GROUP BY rtype.TYP_ID, rtype.TYP_NAME
      ORDER BY COUNT(r.RES_ID) DESC
      ) WHERE ROWNUM <= 3
      """)

    favroomtype = c.fetchall()
    
    roomtype = []
    totalbookingincome = []
    totalbooking = []
    for x in range(len(favroomtype)):
      rtypegrid[x+1,0:3] = Label(str(x+1)+". " + favroomtype[x][1] +", " + str(favroomtype[x][2]))
      roomtype.append(favroomtype[x][1])
      totalbookingincome.append(favroomtype[x][3])
      

    plt.bar(roomtype,totalbookingincome, color= 'turquoise', width = 0.4) 
    plt.bar(roomtype,totalbookingincome, color= 'turquoise', width = 0.4) 

    for x,y in zip(roomtype,totalbookingincome):
        label = "$ {}".format(y)
        plt.annotate(label, 
                    (x,y), 
                    textcoords="offset points", 
                    xytext=(0,2.5), 
                    ha='center')

    plt.ylim(0)
    plt.xlabel("Room Type")
    plt.ylabel("Total Income")
    plt.title("Most Popular Room Type")
    
    #visualize the data
    with favroomtypedashboard:
      plt.show()

    rtypegrid[7:15,0:5] = widgets.HBox([favroomtypedashboard])
  showfavrtype(0)
  rtypegrid[0,0:4] = widgets.HBox([favrtype_label])
  rtypegrid[16,0] = widgets.HBox([frfav_tohome_button])
  display(rtypegrid)
    



############################################################################
################## DAILY ROOM STATUS WITH ROOM TYPE FILTER #################
############################################################################


with tb_all.output_to('View Status'):
  ########################## LABEL ##########################
  # dashboard title
  dailyroomstatus_label = Label('Daily Room Status',bold=True)

  # date label 
  DRStoday = datetime.date.today()
  DRSdate_label = Label("Date :" + DRStoday.strftime("%d %B %Y"))

  DRS_select_status_label = Label('Select Cleaning Status:',bold=True)
  DRS_select_room_type_label = Label("Select Room Type: ", bold = True)


  DRS_select_status_dropdown = widgets.Dropdown(
        options = ['Cleaned', 'Need Cleaning', 'Yet to Check In', 'Checked In', 'Checked Out' ], value = None)

  c.execute ("SELECT TYP_ID, TYP_NAME FROM ROOM_TYPE")

  room_type_dict = {}
  for row in c:
    room_type_dict[row[0]] = row[1]

  DRS_select_room_type_dropdown = widgets.Dropdown(options = room_type_dict.values(), value = None)

  ########################## BUTTONS ###########################
  DRSbackhome_button = Button(description='Home')
  DRSRetrieve_no_button = Button(description='Retrieve Number of Room')
  DRSclear_button = Button(description='Clear')


  ########################## FUNCTION ##########################
  # default chart/daily chart
  def retrieve_cleaning_rooms(i):
    
    if DRS_select_room_type_dropdown.value == None:

      if DRS_select_status_dropdown.value == None:
        #print("Please select a status")
        DailyRoomStatusgrid2[3,2:4] = widgets.HBox([Label('Please select a status')])
        return
      
      elif (DRS_select_status_dropdown.value == "Cleaned") or (DRS_select_status_dropdown.value == "Need Cleaning"):
        c.execute (
          """ 
          SELECT ROM_ID, ROM_NO 
          FROM ROOM 
          WHERE ROM_CLEANING = '{}' 
          """
          .format(DRS_select_status_dropdown.value)
          )
      
        rom_dict = {}

        for row in c:
          rom_dict[row[0]] = row[1]
      else: 
        c.execute ("""SELECT ROM_ID, ROM_NO  FROM RESERVATION 
            FULL OUTER JOIN CONTAINS ON contains.reservation_res_id = reservation.res_id 
            FULL OUTER JOIN ROOM ON room.rom_id = contains.room_rom_id 
            WHERE RES_OCCU_STATUS = '{}' 
            """
            .format(DRS_select_status_dropdown.value)
            )

        rom_dict = {}
      
        for row in c:
          rom_dict[row[0]] = row[1]


      status_print_label = Label(str(len(rom_dict)))
      DRS_print_room_label = Label('No of Rooms:')

      
      DailyRoomStatusgrid2[3,0:2] = widgets.HBox([DRS_print_room_label])
      DailyRoomStatusgrid2[3,2:3] = widgets.HBox([status_print_label])

    else:
      for key, value in room_type_dict.items():
          if DRS_select_room_type_dropdown.value == value:
              room_type_id = key
      if DRS_select_room_type_dropdown.value == None or DRS_select_status_dropdown.value == None :
        DailyRoomStatusgrid2[3,2:4] = widgets.HBox([Label('Please select a status')])
        return
      elif (DRS_select_status_dropdown.value == "Cleaned") or (DRS_select_status_dropdown.value == "Need Cleaning"):
        c.execute (
          """ 
          SELECT ROM_ID, ROM_NO
          FROM ROOM 
          WHERE (ROM_CLEANING = '{}') 
          AND (ROOM_TYPE_TYP_ID = '{}')""".format(DRS_select_status_dropdown.value, room_type_id))
      
        rom_dict = {}

        for row in c:
          rom_dict[row[0]] = row[1]

      else: 
        c.execute ("""SELECT ROM_ID, ROM_NO  FROM RESERVATION 
          FULL OUTER JOIN CONTAINS ON contains.reservation_res_id = reservation.res_id 
          FULL OUTER JOIN ROOM ON room.rom_id = contains.room_rom_id 
          FULL OUTER JOIN ROOM_TYPE ON room_type.typ_id = room.room_type_typ_id
          WHERE (RES_OCCU_STATUS = '{}') AND (typ_id = '{}')""".format(DRS_select_status_dropdown.value, room_type_id))

        rom_dict = {}
      
        for row in c:
          rom_dict[row[0]] = row[1]

      
      status_print_label = Label(str(len(rom_dict))) 
      DRS_print_room_label = Label('No of Rooms:')

      DailyRoomStatusgrid2[3,0:2] = widgets.HBox([DRS_print_room_label])
      DailyRoomStatusgrid2[3,2:3] = widgets.HBox([status_print_label])

    
  DRSRetrieve_no_button.on_click(retrieve_cleaning_rooms)

  def RDS_clear(i):

    DailyRoomStatusgrid2[3,0:2] = Label("")
    DailyRoomStatusgrid2[3,2:3] = Label("")


  DRSclear_button.on_click(RDS_clear)

  # Home
  def frstatus_tohome(i):
    RDS_clear(i)
    showPage('Manager View')
    

  DRSbackhome_button.on_click(frstatus_tohome)

  ########################## GRIDS ##########################
  DailyRoomStatusgrid2 = GridspecLayout(10, 11, height='400px', width='1000px', grid_gap='2px')

  DailyRoomStatusgrid2[0,0:4] = widgets.HBox([dailyroomstatus_label])
  DailyRoomStatusgrid2[1,0:6] = widgets.HBox([DRSdate_label])
  DailyRoomStatusgrid2[2,0:2] = widgets.HBox([DRS_select_status_label])
  DailyRoomStatusgrid2[2,2:4] = widgets.HBox([DRS_select_status_dropdown])
  DailyRoomStatusgrid2[2,5:7] = widgets.HBox([DRS_select_room_type_label])
  DailyRoomStatusgrid2[2,7:10] = widgets.HBox([DRS_select_room_type_dropdown])
  DailyRoomStatusgrid2[4,0:6] = widgets.HBox([DRSRetrieve_no_button])
  DailyRoomStatusgrid2[5,0:6] = widgets.HBox([DRSbackhome_button])

  display(DailyRoomStatusgrid2)







# show Staff Login Page at the beginning
showPage('Staff Login Page')
```
